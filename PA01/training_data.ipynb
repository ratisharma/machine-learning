```python
def get_kmers(bases, k):
    """"
    bases: list, a list of the four bases: ['A','T','G','C'] 
    k: int, kmer value, k<len(bases)
    returns: dict, key: kmer, val: position of kmer in the feature vector
    """"
    kmers = [''.join(p) for p in itertools.product(bases, repeat=k)]
    val = list(xrange(pow(len(bases), k)))
    kmers_pos = dict(zip(kmers, val))
    return kmers_pos
```

```python
def get_kmer_seq(seq, k):
    """"
    seq: str, a gene sequence
    k: int, kmer value
    returns: list, kmer_seq: list of all kmer sequences from the original sequence
    """"
    kmer_seq = []
    seq_list = list(seq)
    for i in xrange(len(seq_list)-k+1):
        bs = ''.join(seq_list[i:i+k])
        kmer_seq.append(bs)
    return kmer_seq
```

```python
def get_seq_vec(sequence, kmers_pos, k):
    kmer_seqs = get_kmer_seq(sequence[:-1], k)
    seq_dict = Counter(kmer_seqs)
    vec = np.zeros((pow(4,k),1))
    for k1 in seq_dict.keys():
        num = kmers_pos[k1]
        vec[num,0] = seq_dict[k1]
    vec_norm = (vec-np.min(vec))/np.max(vec)
    vlist = np.transpose(vec_norm).tolist()
    return pd.Series(vlist)
#     return vlist
```
